```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torchvision
import torchvision.transforms as T

path_to_mnist = 'data\\MNIST'

pallette = [[31, 119, 180],
             [255, 127, 14],
             [44, 160, 44],
             [214, 39, 40],
             [148, 103, 189],
             [140, 86, 75],
             [227, 119, 194],
             [127, 127, 127],
             [188, 189, 34],
             [23, 190, 207]]
```

```python
pretrain_transform = T.Compose([T.Resize(28),T.ToTensor()])
dataset = torchvision.datasets.MNIST('data', train=False, transform=pretrain_transform, target_transform=None, download=False)
data_loader = torch.utils.data.DataLoader(dataset, batch_size=10000,shuffle=False)

for x,y in data_loader:
    break
```

```python
def colored(digit, color_id):
    digit = np.tile(digit[:, :, :, np.newaxis], 3)[0]
    c = pallette[color_id]
    digit[:, :, 0] = digit[:, :, 0] / 255 * c[0]
    digit[:, :, 1] = digit[:, :, 1] / 255 * c[1]
    digit[:, :, 2] = digit[:, :, 2] / 255 * c[2]
    return digit
```

```python
mask = []
for i in range(10):
    mask.append(y==i)
```

```python
fig, ax = plt.subplots(4,6,figsize=(12,8))

for i in range(4):
    for j in range(6):
        img = colored(x[mask[j]][i],i)
        ax[i][j].imshow(img)
        ax[i][j].axis('off')
#plt.tight_layout()
```
